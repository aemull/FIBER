```python
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta

Daftar_Saham = pd.read_excel("database/Daftar Saham.xlsx")
kode_perusahaan = Daftar_Saham['Kode']
kode_perusahaan1 = kode_perusahaan + ".JK"
```

```python
ticker = yf.Tickers(kode_perusahaan1.to_list())
```

```python
data_saham_emitem = ticker.history(period="10y")['Close'].round()
data_saham_emitem.columns = data_saham_emitem.columns.str.replace(r"\.JK$", "", regex=True)
data_saham_emitem
```

```python
tanggal_paling_baru = data_saham_emitem.index.max()
tanggal_paling_lama5 = tanggal_paling_baru - pd.DateOffset(years=5)
tanggal_paling_lama3 = tanggal_paling_baru - pd.DateOffset(years=3)
tanggal_paling_lama1 = tanggal_paling_baru - pd.DateOffset(years=1)
```

```python
harga_sekarang = (data_saham_emitem[data_saham_emitem.index == tanggal_paling_baru]).T
harga_sekarang.columns = ["Now"]
harga_sekarang
```

```python
data_1year = data_saham_emitem[data_saham_emitem.index >= tanggal_paling_lama1]
daftar_perusahaan_1thn = data_1year[data_1year.index.year == tanggal_paling_lama1.year].dropna(axis=1, how="all")
daftar_perusahaan_1thn = daftar_perusahaan_1thn.columns.to_list()
data_1year = data_1year[daftar_perusahaan_1thn]
data_1year
```

```python
data_3year = data_saham_emitem[data_saham_emitem.index >= tanggal_paling_lama3]
daftar_perusahaan_3thn = data_3year[data_3year.index.year == tanggal_paling_lama3.year].dropna(axis=1, how="all")
daftar_perusahaan_3thn = daftar_perusahaan_3thn.columns.to_list()
data_3year = data_3year[daftar_perusahaan_3thn]
data_3year
```

```python
data_5year = data_saham_emitem[data_saham_emitem.index >= tanggal_paling_lama5]
daftar_perusahaan_5thn = data_5year[data_5year.index.year == tanggal_paling_lama5.year].dropna(axis=1, how="all")
daftar_perusahaan_5thn = daftar_perusahaan_5thn.columns.to_list()
data_5year = data_5year[daftar_perusahaan_5thn]
data_5year
```

```python
data_5year_max = data_5year.max().to_frame(name="Max 5 Years")
data_3year_max = data_3year.max().to_frame(name="Max 3 Years")
data_1year_max = data_1year.max().to_frame(name="Max 1 Years")
saham_from_max = pd.concat([data_1year_max,data_3year_max,data_5year_max,harga_sekarang],axis=1)
saham_from_max
```

```python
def check_status_max(row):
    now = row["Now"]
    max_5 = row["Max 5 Years"]
    max_3 = row["Max 3 Years"]
    max_1 = row["Max 1 Years"]

    if pd.notna(max_5) and now < max_5:
        return "Lowest than 5 years"
    elif pd.notna(max_3) and now < max_3:
        return "Lowest than 3 years"
    elif pd.notna(max_1) and now < max_1:
        return "Lowest than 1 years"
    else:
        return None
```

```python
saham_from_max["Status"] = saham_from_max.apply(check_status_max, axis=1)
status_stock_max = saham_from_max["Status"].dropna()
status_stock_max
```

```python
data_5year_min = data_5year.min().to_frame(name="Min 5 Years")
data_3year_min = data_3year.min().to_frame(name="Min 3 Years")
data_1year_min = data_1year.min().to_frame(name="Min 1 Years")
saham_from_min = pd.concat([data_1year_min,data_3year_min,data_5year_min,harga_sekarang],axis=1)
saham_from_min
```

```python
def check_status_min(row):
    now = row["Now"]
    min_5 = row["Min 5 Years"]
    min_3 = row["Min 3 Years"]
    min_1 = row["Min 1 Years"]

    if pd.notna(min_5) and now < min_5:
        return "Lowest than 5 years"
    elif pd.notna(min_3) and now < min_3:
        return "Lowest than 3 years"
    elif pd.notna(min_1) and now < min_1:
        return "Lowest than 1 years"
    else:
        return None
```

```python
saham_from_min["Status"] = saham_from_min.apply(check_status_min, axis=1)
status_stock_min = saham_from_min["Status"].dropna()
status_stock_min 
```

```python
data_5year_mean = data_5year.mean().round().to_frame(name="Mean 5 Years")
data_3year_mean = data_3year.mean().round().to_frame(name="Mean 3 Years")
data_1year_mean = data_1year.mean().round().to_frame(name="Mean 1 Years")
saham_from_mean = pd.concat([data_1year_mean,data_3year_mean,data_5year_mean,harga_sekarang],axis=1)
```

```python
def check_status_mean(row):
    now = row["Now"]
    mean_5 = row["Mean 5 Years"]
    mean_3 = row["Mean 3 Years"]
    mean_1 = row["Mean 1 Years"]

    if pd.notna(mean_5) and now < mean_5:
        return "Lowest than 5 years"
    elif pd.notna(mean_3) and now < mean_3:
        return "Lowest than 3 years"
    elif pd.notna(mean_1) and now < mean_1:
        return "Lowest than 1 years"
    else:
        return None
```

```python
saham_from_mean["Status"] = saham_from_mean.apply(check_status_mean, axis=1)
status_stock_mean = saham_from_mean["Status"].dropna()
status_stock_mean 
```

```python
with pd.ExcelWriter('output/Stock Status.xlsx') as writer:
    status_stock_max.to_excel(writer, sheet_name='Max', index=True)
    status_stock_min.to_excel(writer, sheet_name='Min', index=True)
    status_stock_mean.to_excel(writer, sheet_name='Mean', index=True)
```
